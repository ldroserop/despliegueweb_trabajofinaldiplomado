HISTOGRAMAS

```python
# Defina la lista de columnas para las que desea crear grÃ¡ficos circulares
columns = ['ai_knowledge_rating', 'ai_personal_use_frequency', 'ai_school_use_frequency', 'awareness_of_chat-gpt', 'interest_in_ai_career']

# Recorra las columnas y cree grÃ¡ficos de barras y grÃ¡ficos circulares
for column in columns:

    column_title = column.replace('_', ' ').title()

# Crea una figura con dos subtemas: histograma y grÃ¡fico circular.
    fig, axes = plt.subplots(1, 2, figsize=(14, 6))

# subtema 1: Histograma
    sns.histplot(data=data, x=column, bins=5, color='skyblue', edgecolor='black', ax=axes[0])
    axes[0].set_title(f'Histogram for {column_title}')
    axes[0].set_xlabel(column_title)
    axes[0].set_ylabel('Frequency')

    # subtema 2: diagrama circular
    value_counts = data[column].value_counts()
    custom_palette = sns.color_palette('tab10', n_colors=len(value_counts))
    axes[1].pie(value_counts, labels=value_counts.index, autopct='%1.2f%%', rotatelabels=False, textprops={'fontsize': 12}, colors=custom_palette)
    axes[1].set_title(f'Pie Chart for {column_title}')

    # Ajustar el espacio entre subtramas
    plt.tight_layout()

    # mostrar la trama
    plt.show()
```
